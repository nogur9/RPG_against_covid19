```python
import matplotlib as plt
import subprocess
import pandas as pd
from ipywidgets import widgets
from IPython.display import display, IFrame, HTML, Audio
# from traitlets import traitlets
import functools
import os
import random
import locale
import numpy as np
import time
from IPython.display import clear_output
import string
from IPython.display import FileLink
import datetime
from ipywidgets import Layout, Button, Box, VBox
# expert name
expert_name = 'Chris'
version = '1.1'
age_bin = "aob_50-65"
task = 'AOB'
if age_bin.endswith('F'):
    gender = 'Female'
elif age_bin.endswith('M'):
    gender = 'Male'
else:
    gender = None
# config params

print_id = True
test_run = False
link = True
mandatory_fi_report = False
image = True
python = 1
# global vars
task_data = pd.read_csv(f"resources/DQ.csv")
task_data = task_data.drop_duplicates('taskData.elm_id')
ages = pd.read_excel(r"resources/Age ranges for matching.xlsx")
#age criteria
lower = ages[ages["age bin"] == age_bin[4:]]["selectionAgeBinLow"].values[0]
higher = ages[ages["age bin"] == age_bin[4:]]["selectionAgeBinHigh"].values[0]
task_data = task_data[(task_data['ageV1'] >= lower) & (task_data['ageV1'] < higher)]

if gender is not None:
    task_data = task_data[task_data.gender == gender]
    

reports_path = f"resources/{expert_name}/data/{age_bin}"

subjectScoresView_path = f"resources/{age_bin}/subjectScoresView.xlsx"
subjectScoresView_df = pd.read_excel(subjectScoresView_path)
subjectScoresView_df['userTags.0'] = subjectScoresView_df['userTags.0'].str.extract(r'^([\w\-]+)')

aob_old_output_path = f"saved reports/{expert_name}/{age_bin}_AOB.xlsx"
vgng_old_output_path = f"saved reports/{expert_name}/{age_bin}_VGNG.xlsx"
old_df_aob = pd.read_excel(aob_old_output_path)#.drop(['Unnamed: 0'], axis=1)
cols  = [col for col in df_aob.columns if df_aob[col].dtype is not np.dtype('bool')]
df_aob[cols] = df_aob[cols].replace(np.nan, None)
aob_new_output_path = f"saved reports/{expert_name}/{age_bin}_AOB_redo.xlsx"
vgng_new_output_path = f"saved reports/{expert_name}/{age_bin}_VGNG_redo.xlsx"


#df_star_results = pd.read_csv(f"resources/{age_bin}/star_report.csv")
star_TD = 'taskData.elm_id'
TD = 'taskData._id.$oid'
sysid = r'taskData.acqSysId'
conditions = {'AOB': {'frequent':0, 'novel':1, 'target':2, 'AOB':3}, 'VGNG': {'Go':0, 'NoGo': 1, 'VGNG': 2}}

tests = {'data_quality':1, 'Visible_componants_bna':3, 'Visible_componants_fi':5, 
         'Alpha-waves':7, 'High frequency oscillations':9, 'time_shift':11}

task_tests = {'final desicion': 1, 'signal energy':3, 'major artifacts':5, 'artifacts_effects':7, 'free_text':8}

unique_tests = {'novel':{'miscatch_P3a':13}, 'target':{'miscatch_P3b':13},
    'Go': {'miscatch_P200':13, 'miscatch_P3b':15},
    'NoGo': {'miscatch_N2': 13, 'miscatch_P3a':15}
               }

Visible_componants_bna = {
    'frequent':{'P50_bna':0, 'N1_bna': 1, 'P2_bna':2},
    'novel': {'P50_bna':0, 'N1_bna': 1, 'P3a_bna':2},
    'target':{'P50_bna':0, 'N1_bna': 1, 'P2_bna':2, 'P3b_bna':3},
    'Go': {'P200_bna':0, 'P3b_bna':1,},
    'NoGo': {'N2_bna': 0, 'P3a_bna':1}
        }

Visible_componants_fi = {
    'frequent':{'P50_fi':0, 'N1_fi': 1, 'P2_fi':2},
    'novel': {'P3a_fi':0},
    'target':{'P3b_fi':0},
    
    'Go': {'P200_fi':0, 'P3b_fi':1},
    
    'NoGo': {'N2_fi': 0, 'P3a_fi':1}
        }

reasons_to_fail=['DQ-wise', 'algorithm-wise', 'get second opinion']
artifacts = {'Blink contamination':0,
            'Eye-movement contamination':1, 
            'Bilateral P3a':2,
            'low SNR':3,
            'Noisy baseline':4,
            'Large N2 affecting P3a/b':5,
            'P3a ‘leaked’ to P3b location':6,
            'P3b ‘leaked’ to P3a location':7,
            'P3a awkward distribution':8,
            'P3b awkward distribution':9,
            'Double peak caught in P3a':10,
            'Double peak caught in P3b':11,
            'Old age bin network':12,
            'ERP not normal':13
            }
mandatory_fields = {'AOB':
                    [f'data_quality_{cond}' for cond in ['frequent', 'novel', 'target']],
                    'VGNG': 
                    [f'data_quality_{cond}' for cond in ['go', 'nogo']]}
def task_parser(x):
    if 'AOB' in x:
        return 'AOB'
    else:
        return 'VGNG'
    
def get_columns(cond):
    columns = ['data_quality', 'Alpha-waves', 'High frequency oscillations', 'time_shift'] + list(Visible_componants_bna[cond].keys()) + list(Visible_componants_fi[cond].keys())
    if cond == 'novel':
        columns.append('miscatch_P3a')
    elif cond == 'target':
        columns.append('miscatch_P3b')
    elif cond == 'Go':
        columns.extend(['miscatch_P200', 'miscatch_P3b'])
    elif cond == 'NoGo':
        columns.extend(['miscatch_N2', 'miscatch_P3a'])
    
    columns = [f'{col}_{cond}' for col in columns]
    return columns


frequent_columns = get_columns('frequent')
novel_columns = get_columns('novel')
target_columns = get_columns('target')
AOB_columns = [*[f'{col}_AOB' for col in reasons_to_fail], 'signal energy_AOB', 
               *[f'{col}_AOB' for col in list(artifacts.keys())], 'artifacts_effects_AOB', 'free_text_AOB']

go_columns = get_columns('Go')
nogo_columns = get_columns('NoGo')
VGNG_columns = [*[f'{col}_VGNG' for col in reasons_to_fail], 'signal energy_VGNG', 
                *[f'{col}_VGNG' for col in list(artifacts.keys())], 'artifacts_effects_VGNG', 'free_text_VGNG']

vgng_columns = [TD, 'acqsys', 'time', 'reviewer', 'version', *go_columns, *nogo_columns, *VGNG_columns]
aob_columns = [TD, 'acqsys', 'time', 'reviewer', 'version', *frequent_columns, *novel_columns, *target_columns, *AOB_columns]

task_col = 'subjectScores.virtualSensors.taskData.taskClass.elm_name'
task_data['appended_task_column'] = task    #df_star_results["userTags.0"].apply(task_parser)


def create_binary_buttons(text, options=[None, 'yes', 'no'], button_style=''):
    lb = [widgets.Label(text), 
          widgets.ToggleButtons(options=options, button_style=button_style)]     
    return lb

def data_quality_buttons():
    lb = [widgets.Label('Mark the data quality in general:'), 
          widgets.ToggleButtons(options=[None, 'Unusable', 'Bad', 'Acceptable', 'Good', 'Very Good'], button_style='primary')]        
    return lb

def signal_energy():
    lb = [widgets.Label("Mark the signal energy relative to normal:")] + [widgets.Box([widgets.Label("Flat signal"),
                widgets.IntSlider(min=1, max=5, readout=False), widgets.Label("Very high energy")])]
    return lb


def create_Visible_componants_bna(cond):
    lb = [widgets.Label("Visible components in the BNA report")]+[widgets.Box([widgets.Checkbox(description=i, value=True)
                                                                               for i in Visible_componants_bna[cond].keys()])]
    return lb

def create_Visible_componants_fi(cond):
    lb = [widgets.Label("Visible components in the FI report")]+[widgets.Box([widgets.Checkbox(description=i, value=True)
                                                                              for i in Visible_componants_fi[cond].keys()])]
    return lb

def create_major_artifacts():
    lb = [widgets.Label("Any major artifacts seen in the broad band:")]+[widgets.GridBox([widgets.Checkbox(description= i) for i in 
        artifacts.keys()])]
    return lb


def cearte_reason_to_fail():
    lb = [widgets.Label("Final decision, if the report didn't pass, where did it fail?")]+[widgets.Box([widgets.Checkbox(description= i) for i in 
        reasons_to_fail])]
    return lb


def reverse_assignment_extract_task_results(current_tab, subj_results, task):
    for test in task_tests.keys():
        if test == 'final desicion':
            for dec in reasons_to_fail:
                current_tab.children[task_tests['final desicion']].children[reasons_to_fail.index(dec)].value = bool(subj_results[f'{dec}_{task}'].values[0])
                
        elif test == 'signal energy':
            current_tab.children[task_tests['signal energy']].children[1].value = subj_results[f'{test}_{task}'].values[0]
                        
        elif test == 'major artifacts':
            for art in artifacts.keys():
                try:
                    current_tab.children[task_tests['major artifacts']].children[artifacts[art]].value = bool(subj_results[f'{art}_{task}'].values[0])
                except KeyError:
                    pass
        else:
            #print(current_tab.children[task_tests[test]].value)
            #print(subj_results[f'{test}_{task}'].values[0])
            current_tab.children[task_tests[test]].value = subj_results[f'{test}_{task}'].values[0]
    return current_tab


def reverse_assignment_extract_conditions_results(cond, current_tab, subj_results):
    for test in tests.keys():
        if test == 'Visible_componants_bna':
            for componant in Visible_componants_bna[cond].keys():
                        current_tab.children[tests['Visible_componants_bna']].children[Visible_componants_bna[cond][componant]].value = bool(subj_results[f'{componant}_{cond}'].values[0])
        elif test == 'Visible_componants_fi':
            for componant in Visible_componants_fi[cond].keys():
                        current_tab.children[tests['Visible_componants_fi']].children[Visible_componants_fi[cond][componant]].value = bool(subj_results[f'{componant}_{cond}'].values[0])
        else:
            current_tab.children[tests[test]].value = subj_results[f'{test}_{cond}'].values[0]
                
    if cond in unique_tests.keys():
        for test in unique_tests[cond].keys():
            current_tab.children[unique_tests[cond][test]].value = subj_results[f'{test}_{cond}'].values[0]
            
    return current_tab

def reverse_assignment(oid, subj_acq, children, start_time, task):
    subj_results = old_df_aob[old_df_aob[TD]==oid]
    subj_results = subj_results.where(pd.notnull(subj_results), None)
    for cond in conditions[task].keys():
        
        current_tab = children[conditions[task][cond]]
        
        if cond == task:
            reverse_assignment_extract_task_results(current_tab, subj_results, task)
        else:    
            reverse_assignment_extract_conditions_results(cond, current_tab, subj_results)

    return children

def get_acqSysId(oid):
    try:
        acqsys = task_data[task_data[star_TD] == oid][sysid].unique()[0]
        return acqsys
    except IndexError:
        return -1

def get_ids(oid_iter):  
    while True:
        try:
            oid = next(oid_iter)
            subj_acq = get_acqSysId(oid)
            if subj_acq == -1:
                continue
            fi_exists = os.path.exists(os.path.join(reports_path, f"{subj_acq}_fi.html"))
            task = task_data[task_data[star_TD] == oid]['appended_task_column'].unique()[0]
        except StopIteration:
            print("data saved under your name in:")
            print (r"saved reports")
            print("Thank you")
            raise StopIteration
            
        if mandatory_fi_report:
            if task == 'AOB':
                if (fi_exists) and (oid in list(old_df_aob[TD])) and (oid not in list(new_df_aob[TD])):
                    break
            elif task == 'VGNG':
                if (fi_exists) and (oid in list(old_df_vgng[TD])) and (oid in list(new_df_vgng[TD])):
                    break       
        else:
            print(oid)
            if task == 'AOB':
                if oid in list(old_df_aob[TD]) and (oid not in list(new_df_aob[TD])):
                    break
            elif task == 'VGNG':
                if oid in list(old_df_vgng[TD]) and (oid not in list(new_df_vgng[TD])):
                    break      
    
    return task, oid, subj_acq

def get_reports_substitute(oid):
    
    tmp_df = subjectScoresView_df[subjectScoresView_df['taskData.elm_id.$oid']==oid][['_id.$oid', 'flow.elm_id.$oid', 'userTags.0']]
    htmls = []

    for i in tmp_df.values:
        view_id, flow_id, tag = i
        http_template = fr"https://tehila-wb.dev.mybna.net:3000/browser/subjectScoresView/{view_id}?breadCrumbs=%5B%7B%22name%22%3A%22subjectScoresView%22%2C%22url%22%3A%22/browser%23subjectScoresView/criteria/%7B%20%20%20%20%5C%22flow.elm_id%5C%22%3A%20%7B%20%20%20%20%20%20%20%20%5C%22%24oid%5C%22%3A%20%5C%22{flow_id}%5C%22%20%20%20%20%7D%7D/sortDesc/_id%22%7D%5D"
        htmls.append(HTML(f"""<a href= {http_template} target="_blank">{tag}</a>"""))
        
    return htmls

def open_html_reports(subj_acq):  
    reports = []
    
    if os.path.exists(os.path.join(reports_path, f"{subj_acq}_bna.html")):
        if link:
            reports.append(HTML(f"""<a href="/files/labeling/{reports_path}/{subj_acq}_bna.html" target="_blank">bna file</a>"""))
        else:
            reports.append(IFrame(f"{reports_path}/{subj_acq}_bna.html", width=900, height=800))


    if os.path.exists(os.path.join(reports_path, f"{subj_acq}_fi.html")):
        if link:           
            reports.append(HTML(f"""<a href="/files/labeling/{reports_path}/{subj_acq}_fi.html" target="_blank">fi file</a>"""))
        else:
            reports.append(IFrame(f"{reports_path}/{subj_acq}_fi.html", width=900, height=800))
        
    return reports

def create_VGNG_tab(start_time, oid ,subj_acq, oid_iter):
    tab = widgets.Tab()
    children = []
    
    for i, cond in enumerate(conditions['VGNG'].keys()):
        if cond == 'VGNG':
            #final_score_buttons = create_binary_buttons("final decision for report", options=[None, 'release', 'reject', 'get second opinion'])
            exit_button = widgets.Button(description=f"Done")
            exit_button.on_click(functools.partial(next_report, task='VGNG', start_time=start_time, oid=str(oid),subj_acq=subj_acq, children=children, oid_iter=oid_iter))
            free_text = widgets.Text("Labeler Comments: ")
            reason_to_fail = cearte_reason_to_fail()
            artifacts_effects = create_binary_buttons("if there are artifacts, how much did they affect the capsule?", options=[None, 'no effect', 'some effect', 'significant'], button_style='primary')
            children.append(widgets.GridBox([*reason_to_fail, *signal_energy(), *create_major_artifacts(),
                                             *artifacts_effects,dropdown, free_text, exit_button]))
        else:
            task_widgets = [*data_quality_buttons(),
                            *create_Visible_componants_bna(cond), *create_Visible_componants_fi(cond),
                            *create_binary_buttons('Alpha-waves', options=[None, 'low', 'medium', 'high'], button_style='success'),
                            *create_binary_buttons('High frequency oscillations', options=[None, 'low', 'medium', 'high'], button_style='info'),
                            *create_binary_buttons('time_shift', options=['no', 'yes'])
                           ]
            if cond == 'Go':
                task_widgets.extend(create_binary_buttons("miscatch P200", button_style='warning'))
                task_widgets.extend(create_binary_buttons("miscatch P3b", button_style='warning'))
            elif cond == 'NoGo':
                task_widgets.extend(create_binary_buttons("miscatch N2", button_style='warning'))
                task_widgets.extend(create_binary_buttons("miscatch P3a", button_style='warning'))
        
            children.append(widgets.Layout(task_widgets))
            
        tab.set_title(i, str(cond))
        
    tab.children = children
    return tab
 

def create_AOB_tab(start_time, oid ,subj_acq, oid_iter):
    tab = widgets.Tab()
    children = []
    
    for i, cond in enumerate(conditions['AOB'].keys()):
        if cond == 'AOB':
            #final_score_buttons = create_binary_buttons("final decision for report", options=[None, 'release', 'reject', 'get second opinion'])
            
            exit_button = widgets.Button(description=f"Done")
            exit_button.on_click(functools.partial(next_report, task='AOB', start_time=start_time, oid=str(oid), subj_acq=subj_acq, children=children, oid_iter=oid_iter))
            free_text = widgets.Text("Labeler Comments: ")
            reason_to_fail = cearte_reason_to_fail()
            artifacts_effects = create_binary_buttons("if there are artifacts, how much did they affect the capsule?", options=[None, 'no effect', 'some effect', 'significant'], button_style='primary')
            children.append(widgets.GridBox([*reason_to_fail, *signal_energy(), *create_major_artifacts(),
                                             *artifacts_effects, free_text, exit_button]))
        
        else:
            task_widgets = [*data_quality_buttons(),
                            *create_Visible_componants_bna(cond), *create_Visible_componants_fi(cond),
                            *create_binary_buttons('Alpha-waves', options=[None, 'low', 'medium', 'high'], button_style='success'),
                            *create_binary_buttons('High frequency oscillations', options=[None, 'low', 'medium', 'high'], button_style='info'),
                            *create_binary_buttons('time_shift', options=['no', 'yes'])
                           ]
            if cond == 'novel':
                task_widgets.extend(create_binary_buttons("miscatch P3a", button_style='warning'))
            elif cond == 'target':
                task_widgets.extend(create_binary_buttons("miscatch P3b", button_style='warning'))
        
            children.append(widgets.GridBox(task_widgets))
            
        tab.set_title(i, str(cond))
        
    tab.children = children
    return tab

def get_image():
    rnd_img = random.choice([i for i in os.listdir(f"resources/{expert_name}/images") if i.endswith(".jpg")])
    file = open(os.path.join(f"resources/{expert_name}/images",rnd_img), "rb")
    image = file.read()
    return widgets.Image(
        value=image,
        width=300,
        height=400,
    )


def get_python():
    print("you can test your skills in:\nhttp://192.168.1.118:8888/notebooks/labeling/developers/jupyter_notebook_for_fun.ipynb")
    try:
        global python
        file = open(os.path.join(f"resources", "python", f"{python}.png"), "rb")
        python = python + 1
    except OSError:
        file = open(os.path.join(f"resources", "python", f"last_image.jpg"), "rb")
    image = file.read()
    return widgets.Image(
        value=image,
        width=700,
        height=800,
    )

def get_audio():
    rnd_audio = random.choice([i for i in os.listdir(f"resources/music") if i.endswith(".mp3")])
    return Audio(os.path.join(f"resources/music", rnd_audio))

def run(oid_iter):
    clear_output()
    print("The older BNA/ FI reports may be with wrong age-bins!")
    try:
        task, oid, subj_acq = get_ids(oid_iter)
    except StopIteration:
        return
    
    if print_id:
        print("acqsys id =", subj_acq)
        print("TD id =", oid)
        
    htmls = get_reports_substitute(oid)
    reports = open_html_reports(subj_acq)
    
    start_time = datetime.datetime.now()
    
    if task == 'AOB':
        tab = create_AOB_tab(start_time, oid ,subj_acq, oid_iter)
    else:
        tab = create_VGNG_tab(start_time, oid ,subj_acq, oid_iter)
    
    reverse_assignment(oid, subj_acq, tab.children, start_time, task)
    
    if image:
        img = get_image()
        audio = get_audio()
        #py = get_python()
        display(widgets.GridBox([tab, img]), *htmls, *reports, audio)
    else:
        display(widgets.GridBox([tab]), *htmls, *reports)
    return tab.children


def extract_task_results(current_tab, subj_results, task):
    for test in task_tests.keys():
        if test == 'final desicion':
            for dec in reasons_to_fail:
                subj_results[f'{dec}_{task}'] = current_tab.children[task_tests['final desicion']].children[reasons_to_fail.index(dec)].value
                
        elif test == 'signal energy':
            subj_results[f'{test}_{task}'] = current_tab.children[task_tests['signal energy']].children[1].value
                        
        elif test == 'major artifacts':
            for art in artifacts.keys():
                subj_results[f'{art}_{task}'] = current_tab.children[task_tests['major artifacts']].children[artifacts[art]].value
        
        else:
            subj_results[f'{test}_{task}'] = current_tab.children[task_tests[test]].value
    return subj_results


def extract_conditions_results(cond, current_tab, subj_results):
    for test in tests.keys():
        if test == 'Visible_componants_bna':
            for componant in Visible_componants_bna[cond].keys():
                        subj_results[f'{componant}_{cond}'] = current_tab.children[tests['Visible_componants_bna']].children[Visible_componants_bna[cond][componant]].value
        elif test == 'Visible_componants_fi':
            for componant in Visible_componants_fi[cond].keys():
                        subj_results[f'{componant}_{cond}'] = current_tab.children[tests['Visible_componants_fi']].children[Visible_componants_fi[cond][componant]].value
        else:
            if (not cond =='frequent')and((test=='Alpha-waves') or (test=='High frequency oscillations')):
                print(test, cond, current_tab.children[tests[test]].value)
            subj_results[f'{test}_{cond}'] = current_tab.children[tests[test]].value
                
    if cond in unique_tests.keys():
        for test in unique_tests[cond].keys():
            
            print(test, current_tab.children[unique_tests[cond][test]].value)
            subj_results[f'{test}_{cond}'] = current_tab.children[unique_tests[cond][test]].value
            
    return subj_results

def assign_results (oid, subj_acq, children, start_time, task):
    subj_results = {TD: str(oid), "acqsys":str(subj_acq), 'reviewer': expert_name, 'version': version}
    for cond in conditions[task].keys():
        
        current_tab = children[conditions[task][cond]]
        
        if cond == task:
            subj_results = extract_task_results(current_tab, subj_results, task)
        else:    
            subj_results = extract_conditions_results(cond, current_tab, subj_results)
            
    end_time = datetime.datetime.now()
    delta = end_time - start_time
    subj_results['time'] =  delta.total_seconds()
    return subj_results
    
def check_mandatory_fields(subj_results, task):
    for field in subj_results.keys():
        if field in mandatory_fields[task]:
            if subj_results[field] is None:
                return False
    return True


def next_report(b, task=None, start_time=None, oid=None, subj_acq=None, children=None, oid_iter=None):
    subj_results = assign_results(oid, subj_acq, children, start_time, task)
    can_pass_to_next_report = check_mandatory_fields(subj_results, task)
    if can_pass_to_next_report:
        if task=='AOB':
            new_df_aob.loc[new_df_aob.shape[0]]= subj_results
            new_df_aob.to_excel(aob_new_output_path)
            run(oid_iter)   
            
        elif task == 'VGNG':
            new_df_vgng.loc[df_vgng.shape[0]]= subj_results
            new_df_vgng.to_excel(vgng_new_output_path)
            run(oid_iter)
    else:
        b.button_style='warning'
    
delegated_reports = pd.read_csv(os.path.join(reports_path, "delegated_reports.csv"), header=None)
if test_run:               
    oid_iter = iter(delegated_reports[0].unique()[:2])
else:
    oid_iter = iter(delegated_reports[0].unique())
               

if os.path.exists(aob_new_output_path):
    new_df_aob = pd.read_excel(aob_new_output_path)#.drop(['Unnamed: 0'], axis=1)
else: 
    new_df_aob = pd.DataFrame(columns=aob_columns)
    
if os.path.exists(vgng_new_output_path):
    new_df_vgng = pd.read_excel(vgng_new_output_path)#.drop(['Unnamed: 0'], axis=1)
else: 
    new_df_vgng = pd.DataFrame(columns=vgng_columns)
    


c= run(oid_iter);
```
